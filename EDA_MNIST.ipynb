```python
import torch
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
```

```python
# 1. Load MNIST dataset
transform = transforms.ToTensor()

train_dataset = torchvision.datasets.MNIST(
    root='./data', train=True, download=True, transform=transform
)

test_dataset = torchvision.datasets.MNIST(
    root='./data', train=False, download=True, transform=transform
)

print("âœ… MNIST Data Loaded")
print(f"Training Samples: {len(train_dataset)}")
print(f"Testing Samples: {len(test_dataset)}")
```

```python
# 2. Check shape and data type of a single image
sample_img, sample_label = train_dataset[0]
print("Image Shape:", sample_img.shape)  # [1, 28, 28]
print("Label:", sample_label)
print("Data Type:", type(sample_img))
```

```python
# 3. Visualize sample images
def show_images(dataset, num=6):
    fig, axes = plt.subplots(1, num, figsize=(12, 2))
    for i in range(num):
        img, label = dataset[i]
        axes[i].imshow(img.squeeze(), cmap='gray')
        axes[i].set_title(f'Label: {label}')
        axes[i].axis('off')
    plt.show()

print("ðŸ“Š Sample Images from Dataset:")
show_images(train_dataset)
```

```python
# 4. Plot distribution of digit labels
labels = [label for _, label in train_dataset]
plt.figure(figsize=(8, 4))
plt.hist(labels, bins=10, rwidth=0.8, color='skyblue')
plt.title("Distribution of Digits in MNIST Train Dataset")
plt.xlabel("Digit")
plt.ylabel("Frequency")
plt.grid(axis='y')
plt.show()
```
